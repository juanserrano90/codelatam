```python
def getSubType(name_of_file):
  subtype = name_of_file.split('/')[-1].split('_')[1]
  return int(subtype)
```

```python
def makeCopy(file_path, save_path, file_name, subindice):  
  copy_name = file_name.split('.png')[0]+str(subindice)+'_COPY.png'
  shutil.copyfile(save_path+file_name, copy_name)
```
